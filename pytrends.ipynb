```python
from pytrends.request import TrendReq
pytrends = TrendReq(hl='en-US', tz='120') #tz desplazamiento de zona horaria 
```

```python
import pandas as pd
```

```python
kw_list = ["covid"]
pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='VE', gprop='')
```

```python
pytrends.related_queries() #INTERÉS A LO LARGO DEL TIEMPO
```

```python
pytrends.interest_by_region(resolution='VENEZUELA', inc_low_vol=True, inc_geo_code=False) #INTERÉS POR REGIÓN
```

```python
pytrends.related_topics() #TEMAS RELACIONADOS
```

```python
pytrends.trending_searches(pn='united_states') # tendencias de búsquedas para Venezuela en tiempo real
```

```python
pytrends.top_charts(202103, hl='en-US', tz=300, geo='GLOBAL')
```
